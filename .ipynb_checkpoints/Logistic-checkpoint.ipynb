3、数据预处理

```python
from sklearn import preprocessing  # 数据预处理方法

# 对CASH_OUT与TRANSFER类型数据进行处理（存在欺诈）
used_data = data[(data['type'] == 'TRANSFER') | (data['type'] == 'CASH_OUT')]
used_data.drop(['step', 'nameOrig', 'nameDest', 'isFlaggedFraud'],
               axis = 1,
               inplace = True)
used_data = used_data.reset_index(drop = True)

# 将type转化为类别标签0、1
type_label_encoder = preprocessing.LabelEncoder()
type_category = type_label_encoder.fit_transform(used_data['type'].values)
used_data['typeCategory'] = type_category
used_data.head()
```

```python
sns.heatmap(used_data.corr())  # 变量相关性热力图
```

```python
# 查看转账类型数目
ax = used_data['type'].value_counts().plot(kind = 'bar',
                                           title = 'Transaction Type',
                                           figsize = (6, 6))
for p in ax.patches:  # 标注
    ax.annotate(str(format(int(p.get_height()), 'd')),
                (p.get_x(), p.get_height() * 1.01))
plt.show()

# 查看转账类型中欺诈记录数目
ax = pd.value_counts(
    used_data['isFraud'],
    sort = True
).sort_index().plot(kind = 'bar',
                    title = 'Fraud Transaction Count',
                    figsize = (6, 6))
for p in ax.patches:  # 标注
    ax.annotate(str(format(int(p.get_height()), 'd')),
                (p.get_x(), p.get_height() * 1.01))
plt.show()
```

```python
# 准备数据
feature_names = ['amount',
                 'oldbalanceOrg',
                 'oldbalanceDest',
                 'newbalanceDest',
                 'typeCategory']
X = used_data[feature_names]
y = used_data['isFraud']
print(X.head())
print(y.head())
```

```python
# 解决数据不平衡
num_recoders_fraud = len(used_data[used_data['isFraud'] == 1])       # 欺诈记录数目
fraud_indices = used_data[used_data['isFraud'] == 1].index.values    # 欺诈记录索引
nonfraud_indices = used_data[used_data['isFraud'] == 0].index.values # 非欺诈记录索引

# 随机选取相同数目的非欺诈记录
random_nonfraud_indices = np.random.choice(nonfraud_indices,
                                           num_recoders_fraud,
                                           replace = False)
random_nonfraud_indices = np.array(random_nonfraud_indices)

# 组合两类数据
under_sample_indices = np.concatenate([fraud_indices, random_nonfraud_indices])
under_sample_data = used_data.iloc[under_sample_indices, :]

X_undersample = under_sample_data[feature_names].values
y_undersample = under_sample_data['isFraud'].values
print('欠采样记录：', len(under_sample_data))
```

4、数据建模

```python
# 逻辑斯蒂模型
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

# 模型拟合
X_train, X_test, y_train, y_test = train_test_split(X_undersample,
                                                    y_undersample,
                                                    test_size = 0.3,
                                                    random_state = 0)
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# 预测
y_pred_score = lr_model.predict_proba(X_test)
print(y_pred_score)

fpr, tpr, thresholds = roc_curve(y_test, y_pred_score[:, 1])
roc_auc = auc(fpr, tpr)  # 二分类模型评价指标
print('roc_auc: ', roc_auc)

# 绘制ROC曲线
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([-0.1, 1.0])
plt.ylim([-0.1, 1.01])
plt.xlabel('True Positive Rate')
plt.ylabel('False Position Rate')
plt.show()
```
